```python
import re
import nltk
import string
import numpy as np
import pandas as pd
nltk.download('words')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
df = pd.read_excel('clsvshkr.xlsx') #When arch rivals meet Carlsen vs Hikaru
df[275:285]
```

```python
df['Comment']=df['Comment'].apply(str)
```

```python
words = set(nltk.corpus.words.words())

def rem_ne(text):
  if(text.lower() in words or not text.isalpha()):
    return(text.lower())
  else:
    return("")

rem = lambda x: rem_ne(x)
df['Comment'] = df['Comment'].apply(rem)
df[275:285]
```

```python
TAG_RE = re.compile(r'<a(?s).*>(?s).*<\/a>')

def link_rem(text):
 return TAG_RE.sub('', text)
```

```python
df['Comments'] = df['Comment'].apply(link_rem)
df.drop(columns='Comment',inplace=True)
df[275:285]
```

```python
def clean_text(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    text = text.lower()
    #text = re.sub('[^A-Za-z0-9]','',text)
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    # text = re.sub('<[^>]+>','',text) #'<a(?s).*>(?s).*<\/a>'
    return text

clean = lambda x: clean_text(x)
df['Comments'] = df['Comments'].apply(clean)
df[275:285]
```

```python
df['Comments'] = df['Comments'].str.replace('[^A-Za-z0-9]', ' ', flags=re.UNICODE)
df[275:285]
```

```python
wnl = WordNetLemmatizer()

df['Comments'] = df['Comments'].apply(lambda x: wnl.lemmatize(x))
df[275:285]
```

```python
stop_words = stopwords.words('english')

df['Comments'] = df['Comments'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
df[275:285]
```

```python
df['Comments'].replace('',np.nan,inplace=True)
df[275:285]
```

```python
df.dropna(inplace=True)
df.reset_index(inplace=True,drop=True)
```

```python
df[275:285]
```

```python
find_year = re.compile(r'\d{4}')
def year(text):
  return(re.findall(find_year,text)[0])

df['year'] = pd.to_numeric(df['Time'].apply(year))
df.drop(columns=['Time','Likes','Reply Count'],inplace=True)
```

```python
df.head()
```

```python
df_2020 = df[df['year']==2020]
df_2020.reset_index(inplace = True, drop = True)
df_2020.drop(columns=['year','Name'],inplace=True)

df_2021 = df[df['year']==2021]
df_2021.reset_index(inplace = True, drop = True)
df_2021.drop(columns=['year','Name'],inplace=True)

df_2022 = df[df['year']==2022]
df_2022.reset_index(inplace = True, drop = True)
df_2022.drop(columns=['year','Name'],inplace=True)
```
